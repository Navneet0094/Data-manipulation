```python
df = pd.DataFrame({
    'gender': ['male', 'male', 'female', 'female', 'male', 'female', 'male', 'female'],
    'education_level': ['high school', 'college', 'college', 'graduate', 'high school', 'graduate', 'college', 'graduate'],
    'score': [75, 82, 88, 95, 69, 92, 78, 85]
})
df
```

```python
pd.crosstab(df.gender,df.education_level)
```

```python
pd.crosstab(df.education_level,df.gender)
```

```python
pd.crosstab(df.gender,df.education_level,values = df.score,aggfunc='mean')
```

```python
pd.crosstab(df.gender,df.education_level,normalize = True , margins = True)
```
